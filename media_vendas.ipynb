```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01',
        '2020-02-01',
        '2021-03-01',
        '2022-04-01',
        '2023-05-01',
        '2023-06-01',
        '2020-07-01',
        '2021-08-01',
        '2022-09-01',
        '2023-10-01',
        '2022-11-01',
        '2023-12-01',
        ],
    'valor': [
        400000,
        890000,
        760000,
        430000,
        920000,
        340000,
        800000,
        500000,
        200000,
        900000,
        570000,
        995000,
        ]
}
dict_faturamento
```

```python
df_faturamento=pd.DataFrame(dict_faturamento)
df_faturamento
```

```python
df_faturamento
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento
```

```python
df_faturamento.info()
```

```python
#gráfico faturamento
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))
df_faturamento.plot.bar(x='data_ref', y='valor', ax=ax[0])
ax[0].ticklabel_format(style='plain', axis='y')
ax[0].set_title("Faturamento - Barras")

df_faturamento.plot.line(x='data_ref', y='valor', ax=ax[1])
ax[1].ticklabel_format(style='plain', axis='y')
ax[1].set_title("Faturamento - Linha")
plt.tight_layout()
plt.show()
```
